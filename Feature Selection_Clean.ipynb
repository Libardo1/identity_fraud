```python
import pandas as pd
import numpy as np

data = pd.read_csv('app_with_ev.csv', index_col=0)
data.describe()
```

```python
## subset training,testing, and out of time

data['date'] = pd.to_datetime(data['date'])

#out of time
oot = data.loc[data['date'] >= "11/1/16"]
oot

#training and testing
tnt = data.loc[data['date'] < "11/1/16"]
tnt
```

```python
good = tnt.loc[tnt['fraud'] == 0]
bad = tnt.loc[tnt['fraud'] == 1]

good
```

```python
bad
```

```python
from scipy import stats

ks_list = []
for i in range(80):
    good_list = good[good.columns[i]].tolist()
    bad_list = bad[bad.columns[i]].tolist()
    ks = stats.ks_2samp(good_list, bad_list)
    ks_list.append(ks)
#print(ks_list)
df_ks = pd.DataFrame(np.array(ks_list).reshape(1,160))
df_ks = df_ks.T
df_ks
```

```python
ks_value = df_ks.iloc[::2][0].tolist()
p_value = df_ks.iloc[1::2][0].tolist()
results = pd.DataFrame(data = [ks_value, p_value]).T
results.columns = ['ks value', 'p value']

#sort
results = results.sort_values(['p value','ks value'], ascending=[True, False]).head(40)
#results = results.reset_index()
#results.sort_values(['index'])
results
```

```python
sorted(results.index)
```

```python
def count_fraud(df, c_index):
    new_df = df.iloc[:,[c_index,80]].groupby([df.columns[c_index],'fraud']).size().to_frame(name = 'size').reset_index()
    return new_df

#test function
a_phone_1 = count_fraud(data,0).pivot(index='a_phone_1',columns='fraud',values='size')
a_phone_1
```

```python
import matplotlib.pyplot as plt
#from pylab import rcParams

ax = a_phone_1[[0,1]].plot(kind='bar', title ="a_phone_1", legend=True, fontsize=15)
ax.set_xlabel("occurances", fontsize=20)
ax.set_ylabel("log count of occurances", fontsize=20)
ax.set_yscale('log')

plt.show()
```

```python
def plot(table):
    plt.style.use('seaborn-white')
    ax = table[[0,1]].plot(kind='bar', title = table.index.name, legend=True, fontsize=15)
    ax.set_xlabel("occurances", fontsize=20)
    ax.set_ylabel("log count of occurances", fontsize=20)
    ax.set_yscale('log')
    return(plt.show())
```
